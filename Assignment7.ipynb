```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
Gm = sp.Symbol("G_{m}")
rho = sp.Symbol(r"\rho")
R,T = sp.symbols( "R,T", positive = True)
Me = sp.Symbol("M_{e}")
Gm_eq1 = (rho*R*T)/Me
display(sp.Eq(Gm, Gm_eq1))
print("in units of dyn/cm^2")
display(sp.Eq(Gm, Gm_eq1.subs({rho:0.97,T:25+273, Me : 12000, R:8.314e7})))
```

```python
trep = sp.Symbol(r"\tau_{rep}")
te = sp.Symbol(r"\tau_{e}")
N = sp.Symbol("N")
Ne = sp.Symbol("N_{e}")
M = sp.Symbol("M")
Me = sp.Symbol("M_{e}")
print("In this case using the molar mass ratio is equivalent to N/Ne because division by monomer mass cancels")
display(sp.Eq(trep/te, sp.Pow(N/Ne, 3)))
print("Using 3 as the power law dependence")
display(sp.Eq(trep/te, sp.Pow(N/Ne, 3).subs({N:1e6, Ne:17e3})))
print("Using 3.4 as the power law dependence")
display(sp.Eq(trep/te, sp.Pow(N/Ne, 3.4).subs({N:1e6, Ne:17e3})))
```

How does the width of the plateau region of the stress relaxation modulus of 
entangled polymer solutions (the ratio of reptation time to the Rouse 
time of a strand between entanglements) increase with concentration: 
(i) for athermal solutions; 
(ii) for theta solutions. 

```python
phi = sp.Symbol(r"\phi")
v = sp.Symbol(r"\nu")
print("For Athermal")
display(sp.Eq(trep/te,sp.simplify(sp.Pow(phi,(3*(1-v))/(3*v-1))/sp.Pow(phi,-3*v/(3*v-1)))))
display(sp.Eq(trep/te,sp.simplify(sp.Pow(phi,(3*(1-v))/(3*v-1))/sp.Pow(phi,-3*v/(3*v-1))).subs(v,0.588)))
print("For theta")
display(sp.Eq(trep/te,sp.simplify(sp.Pow(phi,7/3)/sp.Pow(phi,-5/3))))
```
